```python
upset_events = [20161026, 20161111, 20161116, 20161127, 20170128, 20170219, 20170312]
def parse_npdate(array):
    '''
    Input: array of datetime
    Output: array year
            array month
            array day
            array hour
            array min
    '''
    n = len(array)
    year = np.arange(n)
    month = np.arange(n)
    day = np.arange(n)
    hour = np.arange(n)
    mini = np.arange(n)
    for i in range(n):
        (tem1, tem2) = str(array[i][0]).split("T")
        tem1_s = tem1.split("-")
        tem2_s = tem2.split(":")
        year[i] = tem1_s[0]
        month[i] = tem1_s[1]
        day[i] = tem1_s[2]
        hour[i] = tem2_s[0]
        mini[i] = tem2_s[1]
    return year, month, day, hour, mini

index = abs(y_df[:, 0]) > 5
index2 = c2_t[:, 0] < 80
index3 = c3_t[:, 0] < 80 
index2 = index2[0: len(index2) - 1]
index3 = index3[0: len(index3) - 1]
x_ab = x_df[index & (index2 & index3)]
year, month, day, hour, mini = parse_npdate(x_ab)
test = np.unique(year * 10000 + month * 100 + day)
print("If this include all upsets ", check_subset(test, upset_events))
print("What left out", set(upset_events) - set(test))
detected_events = list(set(test))
detected_events.sort()
print(len(detected_events), detected_events)
```
